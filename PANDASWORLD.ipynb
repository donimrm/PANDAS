## **4. Pandas Merging, Joining, and Concatenating**

```python
import numpy as np
import pandas as pd
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
        'A' : ['A0', 'A1', 'A2', 'A3'],
        'B' : ['BO', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K3', 'K4'],
        'C' : ['C0', 'C1', 'C2', 'C3'],
        'D' : ['DO', 'D1', 'D2', 'D3']})
right
```

```python
pd.merge(left, right, on = 'key')
```

```python
pd.merge(left, right,how = 'outer', on = 'key')
```

```python
pd.merge(left, right,how = 'left', on = 'key')
```

```python
pd.merge(left, right, on = 'key')
```

```python
pd.concat([left,right])
```

```python
pd.concat([left,right], axis = 1)
```

```python
left = pd.DataFrame({
        'A' : ['A0', 'A1', 'A2', 'A3'],
        'B' : ['BO', 'B1', 'B2', 'B3']},
        index = ['K0', 'K1', 'K2', 'K3'])
left
```

```python
right = pd.DataFrame({'C' : ['C0', 'C1', 'C2', 'C3'],
                      'D' : ['DO', 'D1', 'D2', 'D3']},
                     index = ['K0', 'K1', 'K3', 'K4'])
right
```

```python
left.join(right)
```

```python
right.join(left)
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K3', 'K4'],
        'C' : ['C0', 'C1', 'C2', 'C3'],
        'E' : ['DO', 'D1', 'D2', 'D3']})
right
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K3', 'K4'],
        'A' : ['A0', 'A1', 'A2', 'A3'],
        'B' : ['BO', 'B1', 'B2', 'B3']})
left
```

```python
right.merge(left)
```

```python
pd.concat([right,left], axis = 1)
```

```python
#Function to extract the car first name and create a new column called car_first_name
def extract_car_first_name(df):
    df['car_first_name'] = df['name'].str.split(' ').str.get(0)    
    return df

#Function to add my_name after the car_first_name and create a new column called car_and_name
def add_car_my_name(df, my_name = None):
    df['car_and_name'] = df['car_first_name'] + my_name    

mpg.pipe(extract_car_first_name).pipe(add_car_my_name, my_name = ' Keren')
    
mpg.head()
```
